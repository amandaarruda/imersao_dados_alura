```python
import pandas as pd
```

```python
url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url)
```

## **Desafio 1: Realizar a média da metragem para cada um dos bairros**

```python
dados.groupby("Bairro").Metragem.mean()
```

```python
#sobre o groupby: A groupby operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups.
#ou ainda: 
dados.groupby("Bairro")["Metragem"].mean()
```

# **Desafio 3: Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, assim como tentar colocar alguma conclusão**

```python
#fazendo uma variável para o número de quartos
n_quartos = dados.Quartos.value_counts()
n_quartos.plot.bar()
```

```python
#fazendo uma variável para o número de banheiros
n_banheiros = dados.Banheiros.value_counts()
n_banheiros.plot.bar()
```

```python
#fazendo uma variável para o número de vagas
n_vagas = dados.Vagas.value_counts()
n_vagas.plot.bar()
```

```python
#fazendo uma variável para as 15 ruas mais frequentes
n_ruas = dados.Rua.value_counts()
n_ruas.head(15).plot.bar()
```

```python
#fazendo uma variável para os preços
n_preço = dados.Valor.value_counts()
n_preço.plot.pie()
```

```python
#ficou um pouco confuso 
#pegando os  preços mais recorrentes:
n_preço.head(10).plot.pie()
```

```python
n_preço.head(10).plot.bar()
```

```python
#transformando a metragem em um gráfico de pizza
n_metragem = dados.Metragem.value_counts()
n_metragem.plot.pie()
```

```python
n_metragem.head(15).plot.pie()
```

```python
#relação entre bairro e metragem
metragem_bairro = dados.loc[:,["Metragem","Bairro"]].groupby("Bairro").mean().sort_values("Metragem",ascending = False)
metragem_bairro.head(10).plot.barh()
```
